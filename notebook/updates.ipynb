# Update documents and collections

```python
from connect import db
#Obtener un libro tirulado La ciudad Verde con todos sus elementos
print(db.libros.find_one({"titulo":"La ciudad Verde"}))
```

```python
from connect import db
from pymongo import UpdateOne

modify = db.libros

# Actualizar un solo documento
res = modify.update_one(
    {"titulo": "La ciudad Verde"},
    {"$inc": {"stock": 10}} # Incrementar el stock en 10
)
print(f"Cambios realizados: {res.matched_count},'total' {res.modified_count}")
```

```python
# restar la cantidad de stock del mismo libro
res = modify.update_one(
    {"titulo": "La ciudad Verde"},
    {"$inc": {"stock": -2}} # Decrementar el stock en 2
)

print(db.libros.find_one({"titulo":"La ciudad Verde"}))
# Solo mostrar el titulo, autor y stock
print(db.libros.find_one({"titulo":"La ciudad Verde"}, {"_id":0, "titulo":1, "autor":1, "stock":1}))
print(f"Cambios realizados: {res.matched_count},'total' {res.modified_count}")
```

```python
# cuantos libros hay del libro la ciudad verde
for doc in modify.find({"titulo": "La ciudad Verde"}):
    print(doc["titulo"], " hay ", doc["stock"] , " libros en stock")
```

```python
# actualizar el stock actual
res = modify.update_one(
    {"titulo": "La ciudad Verde"},
    {"$set": {"stock": 15}} # Establecer el stock a 15
)
# cuantos libros hay del libro la ciudad verde
for doc in modify.find({"titulo": "La ciudad Verde"}):
    print(doc["titulo"], " hay ", doc["stock"] , " libros en stock")
```
