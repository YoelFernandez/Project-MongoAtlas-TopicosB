# Insertando datos

```python
from connect import db

# el primero es la bd y el segundo es la coleccion
libros = db.libros
libro = {
    "titulo": "La ciudad Verde",
    "autor": "Mario Vargas Llosa",
    "anio": 1965,
    "categoria": ["Novela", "Peruana"],
    "stock": 4
}

#Inserta un documento (fila)
res = libros.insert_one(libro)
print("ID del libro insertado:", res.inserted_id)
```
