# Realizando consultas

```python
from connect import db
result = db.libros
# consultar todo
for doc in result.find():
    print(doc)
```

```python
# consulta por autor
for doc in result.find({"autor": "Mario Vargas Llosa"}):
    print(doc["titulo"], "-", doc["autor"], doc["anio"])
```

```python
# libros mayores a 1960
for doc in result.find({"anio": {"$gt": 1960}}): #gt significa greater than
    print(doc["titulo"], "-", doc["autor"], doc["anio"])
```

```python
# Listar libros de ciencia ficcion
for doc in result.find({"generos": {"$gte": "Ciencia Ficcion"}}): #gte significa greater than or equal e incluye "Ciencia Ficcion" en español significa mayor o igual que
    print(doc["titulo"], doc["anio"])
```

```python
# libros con stock mayor a 2
for doc in result.find({"stock": {"$gt": 2}}):
    print(doc["titulo"], "-", doc["stock"])
```
